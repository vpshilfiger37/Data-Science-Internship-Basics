```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
data = {
    'Name': ['Adi', 'Thomas', 'Shiva', 'David', 'Solomon'],
    'Age': [25, 30, 22, 35, 28],
    'Score': [85, 90, 97, 88, 95]
}

df = pd.DataFrame(data)

# Display the first few rows
print("First few rows of the dataset:")
print(df.head())

# Generate basic statistical summaries
print("\nDataset Information:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

# Visualizing data distributions
plt.figure(figsize=(10, 5))

df.hist(figsize=(10, 5), bins=5, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Data', fontsize=14)
plt.show()

# Box plots to detect outliers
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Age', 'Score']], palette='Set2')
plt.title('Box Plot for Age and Score')
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Age', 'Score']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
