Data Loading

```python
import pandas as pd
import seaborn as sns

# Load Titanic dataset from seaborn
df = sns.load_dataset('titanic')
```

Data Cleaning

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# Drop rows with missing values (or you can impute based on context)
df_cleaned = df.dropna()

# Check for duplicates and remove them if any
df_cleaned = df_cleaned.drop_duplicates()
```

Basic Statistics

```python
# Generate descriptive statistics
statistics = df_cleaned.describe()
print(statistics)
```

Visual Analysis

```python
import matplotlib.pyplot as plt

# Bar chart to show survival count by class
sns.countplot(x='pclass', hue='survived', data=df_cleaned)
plt.title('Survival Count by Class')
plt.show()
```

```python
# Pie chart to show gender distribution
gender_count = df_cleaned['sex'].value_counts()
gender_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, title='Gender Distribution')
plt.ylabel('')  # Remove the y-label
plt.show()
```

```python
# Histogram to show age distribution
plt.hist(df_cleaned['age'], bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Keep only numeric columns
df_numeric = df_cleaned.select_dtypes(include=['number'])

# Compute correlation
correlation_matrix = df_numeric.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
