Basic Data Visualization

 Write a Python script (data_visualization.py) that:
- Generates random data and plots a line graph using Matplotlib.
- Saves the plot as an image (plot.png).

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
x = np.arange(0, 10, 0.1)
y = np.sin(x) + np.random.normal(0, 0.1, len(x))

# Plot the line graph
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Random Data', color='blue')
plt.title('Line Graph of Random Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Save the plot as an image
plt.savefig('plot.png')
plt.show()

print("Plot saved as plot.png!")
```
